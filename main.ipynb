```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# read image
img = cv2.imread("input.jpg")

# convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print(gray.shape)

# Laplacian kernel
kernel = np.array([[0,  1, 0],
                   [1, -4, 1],
                   [0,  1, 0]], dtype=np.float32)

# convolution
edges = cv2.filter2D(gray, ddepth=cv2.CV_32F, kernel=kernel)

# normalize to 0255
edges = np.abs(edges)
edges = (255 * edges / edges.max()).astype(np.uint8)

# display
plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.show()
```
